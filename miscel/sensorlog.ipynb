```python
import os 
import json
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
directory_path = "/home/ava/Desktop/phone/phone up"
file_name = os.listdir(directory_path)[0]
file_path = directory_path + "/" + file_name
```

```python
def read_file(file_path):
    quaternions = []
    accelerations = []

    with open(file_path, 'r') as json_file:
        datas = json.load(json_file)
        for data in datas:
            quaternion = (float(data["motionQuaternionX"]), float(data["motionQuaternionY"]),
                        float(data["motionQuaternionZ"]), float(data["motionQuaternionW"]))
            quaternions.append(quaternion)
            
            acceleration = (float(data["motionUserAccelerationX"]), float(data["motionUserAccelerationY"]),
                            float(data["motionUserAccelerationZ"]))
            accelerations.append(acceleration)

    return quaternions, accelerations
```

```python
quaternions, accelerations = read_file(file_path)
```

```python
def plot_list(data):
    data_x = [d[0] for d in data]
    data_y = [d[1] for d in data]
    data_z = [d[2] for d in data]
    
    # Create scatter plots for each data component
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 3, 1)
    plt.scatter(range(len(data_x)), data_x, c='b', marker='o')
    plt.xlabel('Data Point')
    plt.ylabel('Data X')
    plt.title('Data X Scatter Plot')
    
    plt.subplot(1, 3, 2)
    plt.scatter(range(len(data_y)), data_y, c='g', marker='o')
    plt.xlabel('Data Point')
    plt.ylabel('Data Y')
    plt.title('Data Y Scatter Plot')
    
    plt.subplot(1, 3, 3)
    plt.scatter(range(len(data_z)), data_z, c='r', marker='o')
    plt.xlabel('Data Point')
    plt.ylabel('Data Z')
    plt.title('Data Z Scatter Plot')
    
    plt.tight_layout()
    plt.show()
```

```python
plot_list(accelerations)
```

```python
directory_path2 = "/home/ava/Desktop/phone/phone down"
file_name2 = os.listdir(directory_path2)[0]
file_path2 = directory_path2 + "/" + file_name2
quaternions2, accelerations2 = read_file(file_path2)
plot_list(accelerations2)
```
